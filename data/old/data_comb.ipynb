```python
new_df["result"] = new_df["team2score"] > new_df["team1score"]
new_df["result"] = new_df["result"].astype(int)

new_df["wteam"] = new_df["team1"][new_df["result"] == 0]
new_df.loc[new_df["result"] == 1, "wteam"] = new_df["team2"][new_df["result"] == 1]

new_df["lteam"] = new_df["team1"][new_df["result"] == 1]
new_df.loc[new_df["result"] == 0, "lteam"] = new_df["team2"][new_df["result"] == 0]

new_df["wscore"] = new_df["team1score"][new_df["result"] == 0]
new_df.loc[new_df["result"] == 1, "wscore"] = new_df["team2score"][new_df["result"] == 1]

new_df["lscore"] = new_df["team1score"][new_df["result"] == 1]
new_df.loc[new_df["result"] == 0, "lscore"] = new_df["team2score"][new_df["result"] == 0]
new_df["lscore"] = new_df["lscore"].astype(int)
new_df["wscore"] = new_df["wscore"].astype(int)
new_df
```

```python
merged_df = pd.merge(old_df.iloc[::-1], new_df, how="outer", left_on=["YEAR", "ROUND", "WSCORE", "LSCORE"], right_on=["year", "round", "wscore", "lscore"])
```
