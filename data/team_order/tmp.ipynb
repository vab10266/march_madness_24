```python
# Importing the required modules 
import os
import sys
import pandas as pd
from bs4 import BeautifulSoup

# path = 'tmp.html'
def parse_new_html(year):
    path = f'{year}.html'
    print(path)
    soup = BeautifulSoup(open(path),'html.parser')

    games = [x[70:].replace('<', '>') for x in str(soup).split('FINAL') if x[0] == '\n']
    team1 = [x.split('>')[9] for x in games]
    team1score = [x.split('>')[13] for x in games]
    team1seed = [x.split('>')[5] for x in games]
    team2 = [x.split('>')[27] for x in games]
    team2score = [x.split('>')[31] for x in games]
    team2seed = [x.split('>')[23] for x in games]

    df = pd.DataFrame(data = [team1, team1score, team1seed, team2, team2score, team2seed]).T
    df.columns = ['team1', 'team1score', 'team1seed', 'team2', 'team2score', 'team2seed']
    df['result'] = (df['team1score'] < df['team2score']).astype(int)
    df['year'] = year
    if year == 2021:
        df = pd.concat((df.iloc[:6], pd.DataFrame([['Oregon', 75, 7, 'VCU', 70, 10, 0, 2021]], columns=['team1', 'team1score', 'team1seed', 'team2', 'team2score', 'team2seed', 'result', 'year']), df.iloc[6:]), axis=0)
    

    rounds = [[1]*8, [2]*4, [3]*2, [4]*1, [1]*8, [2]*4, [3]*2, [4]*1, [5]*1, [6]*1, [5]*1, [1]*8, [2]*4, [3]*2, [4]*1, [1]*8, [2]*4, [3]*2, [4]*1, [0]*4]
    a = []
    for round in rounds:
        for game in round:
            a.append(game)
    df['round'] = a
    # print(df)
    int_cols = ['team1score', 'team1seed', 'team2score', 'team2seed',
       'result', 'year', 'round']
    df[int_cols] = df[int_cols].astype(int)
    return df
```

```python
all_df = None
for i in range(8):
    if i == 4:
        continue
    df = parse_new_html(2016 + i)
    all_df = pd.concat((all_df, df), axis=0)
all_df
```
